# Section 2: Vanishing Gradients


The vanishing gradient problem is a critical issue that affects the training of deep neural networks, including Multilayer Perceptrons (MLPs). This problem occurs when the gradients of the loss function diminish as they are propagated backward through the network during training. When gradients become too small, they fail to update the weights of the earlier layers effectively, causing the network to stop learning. In this section, we will explore the causes of vanishing gradients, the activation functions that exacerbate the problem, and modern solutions to mitigate this issue.

# Section 3: Backpropagation Optimizations


Backpropagation is the core algorithm used for training neural networks, including Multilayer Perceptrons (MLPs). It calculates the gradients of the loss function with respect to each weight in the network and updates the weights accordingly to minimize the loss. While backpropagation is fundamental, it can be computationally expensive and challenging to optimize, especially for deep networks where issues like slow convergence, gradient explosion, and vanishing gradients can arise.

This section will explore various optimizations and techniques that improve the efficiency and effectiveness of backpropagation. We will focus on modern optimizers such as Adam and RMSprop, learning rate schedules, gradient clipping, and other strategies that help improve the training process in deep MLPs.

## 3.4 Gradient Clipping: Preventing Gradient Explosion


While the vanishing gradient problem is a major issue in deep networks, gradient explosion can also occur, particularly in recurrent networks or networks with many layers. When gradients become too large during backpropagation, they can cause weight updates to be too drastic, leading to instability and divergence.

Gradient clipping is a technique used to prevent gradient explosion by capping the gradients at a predefined maximum value.



### 3.4.1 How Gradient Clipping Works


During backpropagation, if any gradient exceeds the predefined threshold, it is scaled down to that threshold. This ensures that no gradient update is too large, preventing the optimizer from making erratic weight changes.
