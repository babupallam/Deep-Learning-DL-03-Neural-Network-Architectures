# Section 2: Batch Normalization


Batch normalization is a technique that addresses two key issues in training deep neural networks: unstable learning and slow convergence. By normalizing the activations of each layer in a neural network, batch normalization ensures that the input to each layer remains well-behaved, speeding up the training process and improving model stability. In this section, we will delve into the mechanics of batch normalization, explore its benefits, and discuss its role in modern neural network architectures.



## 2.1 The Need for Normalization in Deep Networks


In deep neural networks, particularly those with multiple layers, the distribution of activations can shift as the model trains. This phenomenon is known as internal covariate shift, where the distribution of each layer's inputs changes due to updates in the preceding layers during training. This makes it difficult for the model to learn effectively, as the parameters in each layer are constantly adapting to these shifting inputs.



image.png

## 2.2 How Batch Normalization Works


Batch normalization works by normalizing the output of a given layer so that it has a mean of zero and a standard deviation of one, which stabilizes the learning process and reduces sensitivity to weight initialization.
